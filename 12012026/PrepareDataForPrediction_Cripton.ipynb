```python
cols_to_avg = [
    'cnt_moves','cnt_platform_out_moves','cnt_red','cnt_kruising',
    'sum_pri_delay','cnt_pri_delay','cnt_change_comp','cnt_trains',
    'cnt_S_NORMAL','cnt_S_WRONG','cnt_signals','incident'
]

# === Compute mean per [TCC, datetime] ===
complx_april_TCC_datetime_mean = (
    complx_april.groupby(['TCC', 'datetime'])[cols_to_avg]
      .mean()
      .reset_index()
)

# Add suffix to make meaning explicit
complx_april_TCC_datetime_mean = complx_april_TCC_datetime_mean.rename(
    columns={col: col + "_mean" for col in cols_to_avg}
)
```

```python
complx_april_TCC_datetime_mean['density_mean'] = (complx_april_TCC_datetime_mean['cnt_moves_mean'] 
                                                  / complx_april_TCC_datetime_mean['cnt_S_NORMAL_mean']).fillna(0)

complx_april_TCC_datetime_mean['complexity_mean'] = (complx_april_TCC_datetime_mean['cnt_kruising_mean'] 
                                                     / complx_april_TCC_datetime_mean['cnt_platform_out_moves_mean']).fillna(0)

complx_april_TCC_datetime_mean['non_robustness_mean'] = (complx_april_TCC_datetime_mean['cnt_red_mean'] 
                                                         / complx_april_TCC_datetime_mean['cnt_moves_mean']).fillna(0)

complx_april_TCC_datetime_mean['operational_factor_mean'] = (complx_april_TCC_datetime_mean['cnt_change_comp_mean'] 
                                                             / complx_april_TCC_datetime_mean['cnt_platform_out_moves_mean']).fillna(0)

complx_april_TCC_datetime_mean = complx_april_TCC_datetime_mean.replace([np.inf, -np.inf], 0)
```

```python
# -------- CONFIG ----------
# base network features (without "_mean" suffix in final names)
base_features = [
    'cnt_signals', 'incident', 'density',
    'complexity', 'non_robustness', 'operational_factor'
]

# source table: must contain columns 'TCC', 'datetime', and e.g. 'cnt_signals_mean', ...
src = complx_april_TCC_datetime_mean  # your input

# -------- BUILD NETWORK WIDE TABLE ----------
# ensure datetime is proper dtype
src = src.copy()
src['datetime'] = pd.to_datetime(src['datetime'])

# list of column names in source for these features (with _mean suffix)
src_feature_cols = [f"{f}_mean" for f in base_features]

# keep only columns we need to save memory
mean_tbl = src[['TCC', 'datetime'] + src_feature_cols].copy()

# pivot: (datetime, TCC) -> wide. The result has MultiIndex columns: (feature_mean, TCC)
complx_network = (
    mean_tbl
    .set_index(['datetime', 'TCC'])[src_feature_cols]
    .unstack(level='TCC')   # columns: (feature_mean, TCC)
)

# flatten columns to "<TCC>_<feature>_mean"
network_cols = []
for feature_name, tcc in complx_network.columns:
    # feature_name is like 'cnt_signals_mean'
    # build colname "<TCC>_<feature_name>"
    network_cols.append(f"{tcc}_{feature_name}")

complx_network.columns = network_cols

# reset index so there is ONE row per datetime
complx_network = complx_network.reset_index()

# optional: reduce memory use (float32)
for c in complx_network.columns:
    if c != 'datetime' and pd.api.types.is_float_dtype(complx_network[c]):
        complx_network[c] = complx_network[c].astype('float32')

# fill missing with 0 (your earlier preference)
complx_network = complx_network.fillna(0.0)
complx_network = complx_network.replace([np.inf, -np.inf], 0)
# result:
# complx_network is a dataframe with columns:
# 'datetime', 'T_cnt_signals_mean', 'C_cnt_signals_mean', ..., 'T_incident_mean', ...
# and exactly one row per datetime (all datetimes observed in the source mean table).
```

```python
from datetime import datetime
start_timestamp = pd.Timestamp(datetime(2023, 4, 1, 0, 0, 0), tz='Europe/Paris')

MINUTE= 5

def Append_t(df_time_column):
    """
    Converts the input timestamps to simplified time units where each 5-minute interval is considered a unit.

    Args:
    - df_time_column: pandas Series, the column containing timestamps in ISO 8601 format.

    Returns:
    -pandas Series: A Series containing the converted values of the timestamp column.

    """

    # Calculate the difference in minutes from the starting timestamp
    diff_in_minutes_from_start_timestamp= ((df_time_column - start_timestamp).dt.total_seconds() / 60).round(2)

    # Divide the difference by 5 to get the running number in minutes
    current_minute= (diff_in_minutes_from_start_timestamp// MINUTE).astype('int32') 

    return current_minute



complx_network['t'] = Append_t(complx_network['datetime'])
```

```python
complx_dict = {}

for (tcc, ws), g in complx_april.groupby(["TCC", "WS_NO"]):

    # Sort by datetime to support interval search + nearest search
    g_sorted = g.sort_values("datetime").reset_index(drop=True)

    # Store in dictionary
    complx_dict[(tcc, ws)] = g_sorted

print("complx_dict created:", len(complx_dict), "groups")
```

```python
# --- Build dictionary for network lookup by t ---
network_dict = {
    row["t"]: row.drop("t").to_dict()
    for _, row in complx_network.iterrows()
}

# Sorted list of all available t values in the network table
network_t_values = sorted(network_dict.keys())
network_t_np = np.array(network_t_values)
```

```python
out = []
N = len(s_r_w_df)

# -----------------------------
# LOCAL columns (use TCC + WS_NO)
# -----------------------------
local_cols = [
    'cnt_moves','cnt_platform_out_moves','cnt_red','cnt_kruising',
    'sum_pri_delay','cnt_pri_delay','cnt_change_comp','cnt_trains',
    'cnt_S_NORMAL','cnt_S_WRONG','cnt_signals','incident','density',
    'complexity','non_robustness','operational_factor'
]

# -----------------------------
# NETWORK columns (use Micro_time only)
# -----------------------------
# network_cols = new_cols   # already provided by you

all_cols = local_cols + network_cols


for i, row in s_r_w_df.iterrows():

    tcc = row['TCC']
    ws  = row['WS_NO']
    mt  = row['Micro_time'] - pd.Timedelta(minutes=15)
    t_val = row["t"] - 3
    
    attach = {c: pd.NA for c in all_cols}   # initialize

    # =====================================================
    # 1. LOCAL LOOKUP (TCC + WS_NO + micro_time)
    # =====================================================
    if not (pd.isna(tcc) or pd.isna(ws) or pd.isna(mt)):

        key = (tcc, ws)

        if key in complx_dict:
            g = complx_dict[key]

            # interval match
            mask = (g['datetime'] <= mt) & (mt < g['end_time'])
            if mask.any():
                matched = g.loc[mask.idxmax()]
            else:
                # nearest time match
                idx_min = (g['datetime'] - mt).abs().idxmin()
                matched = g.loc[idx_min]

            # fill local columns
            for c in local_cols:
                attach[c] = matched[c]

    # =====================================================
    # 2. NETWORK LOOKUP (MICRO_TIME ONLY)
    # =====================================================
    # Using the wide lookup table you constructed: "network_dict"
    # Format: network_dict[datetime] ‚Üí row with all TCC_*_mean columns


    if pd.notna(t_val):

        # --- FLOOR NEAREST MATCH ---
        # choose the largest network t <= current t
        mask = network_t_np <= t_val

        if mask.any():
            # take the maximum t <= current t
            nearest_t = network_t_np[mask].max()
        else:
            # no t <= current t ‚Üí take smallest t
            nearest_t = network_t_np[0]

        # Extract network features row
        netrow = network_dict[nearest_t]
        for c in network_cols:
            attach[c] = netrow[c]
            
    # Combine original + network columns
    out.append({**row.to_dict(), **attach})

    if (i+1) % 10000 == 0:
        print(f"{i+1}/{N} processed...")


result_df = pd.DataFrame(out)
print("Complete.")
```
